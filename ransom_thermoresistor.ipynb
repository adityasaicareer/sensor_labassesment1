```python
output=pd.read_csv("resist_output.csv",names=['res','temp'])
```

```python
output
```

```python
resistance=[]
switchon=[]
for i in range(output.shape[0]):
    if(output['temp'].loc[i]>308):
        switchon.append(output['temp'].loc[i])
        resistance.append(output['res'].loc[i])
```

```python
plt.figure(figsize=(5,5),dpi=1000)
plt.plot([i for i in range(60)],output['temp'].loc[1:])
plt.title("Temparature(thermoresistance)")
plt.xlabel('Temp(K)')
plt.ylabel('sample(n)')
```

```python
resistance=np.array(resistance)
```

```python
switchon=np.array(switchon)
```

```python
plt.scatter(resistance,switchon)
```

```python
#for thermo aqustic sensor

aq=pd.read_csv('/Users/chowdaryadithyasaividivada/Documents/22BCT0087-LA1/aqustic.csv',names=['temp'])
```

```python
aq
```

```python
swon=[]
fon=[]
for i in range(aq.shape[0]):
    if(aq['temp'].loc[i]>35):
        swon.append(i)
    if(aq['temp'].loc[i]<6):
        fon.append(i)

swar=np.array(swon)
plt.figure(figsize=(5,3),dpi=1000)
plt.plot(aq['temp'],label='temparature')
plt.scatter(swar,aq['temp'].loc[swon])
print(fon)
plt.scatter(np.array(fon),aq['temp'].loc[fon],color='red',marker='*',label='Fan On')
plt.title("Temparature Graph",fontdict={'fontsize':12,'color':'red'})
plt.xlabel('Sample(n)')
plt.ylabel('Temp(celcius)')
plt.legend()
plt.savefig('/Users/chowdaryadithyasaividivada/Documents/22BCT0087-LA1/aqustic.png')
```
